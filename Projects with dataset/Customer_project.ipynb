# Here we can how flow purchase trend accross seasons

```python
pivot_count = pd.pivot_table(
    df,
    index="Season",
    columns="Category",
    values="INR",
    aggfunc="mean"
)

pivot_count.plot(marker='o')
plt.title("Category-wise Purchase Trend Across Seasons")
plt.xlabel("Season")
plt.ylabel("Number of Purchases")
plt.grid(True)
plt.tight_layout()
plt.show()
```

# age is divied into bins means age group and give their names according age

```python
df["Age_Group"] = pd.cut(df["Age"],bins=[0, 18, 30, 45, 60, 100],labels=["Teen", "Young", "Adult", "MidAge", "Senior"])
```

```python
pivot_count = pd.pivot_table(df,index="Age_Group",columns="Category",values="INR",aggfunc="count")

pivot_count.plot(kind="bar")
plt.title("Category-wise Purchase Count Across Age")
plt.xlabel("Age")
plt.ylabel("Number of Purchases")
plt.grid(True)
plt.tight_layout()
plt.show()
```
